
# create 2 matrices with size 12 and 12, 
# add values
# through slicing get an optional row, column
# get values which are bigger than 5

```python
# 2), 3)
matrix_1 = []
for i in range(0,12):
    matrix_1.append([])
    for j in range(0,12):
        matrix_1[i].append(j + i)
    print(matrix_1[i])
            
```

```python
matrix_2 = []
for i in range(0,12):
    matrix_2.append([])
    for j in range(0,12):
        if i == 0:
            matrix_2[i].append(matrix_1[i][j])
        elif j == 0:
            matrix_2[i].append(matrix_1[i][j])
        else:
            matrix_2[i].append(j + 2*i)
    print(matrix_2[i])
```

```python
for i in matrix_1:
    for j in range(0,12):
        if i[j] > 5:
            print(i[j])
```
